## K-Means

<p style='text-align: justify;'>El algoritmo de K-Means es un algoritmo de Clustering basado en Centroides que agrupa diversos elementos en k grupos basndose en la similitud de las caractersticas de los elementos. El agrupamiento se realiza a travs de un proceso iterativo donde en cada una de sus repeticiones se pretende minimizar la suma del cuadrado de las distancias entre cada uno de los objetos y el centroide de su grupo o clster. El algoritmo consta de tres pasos:</p>

1) Asignacin inicial de los centroides. 

2) Creacin de los conjuntos tomando en cuenta que elementos estn ms cerca a cada centroide.

3) Actualizacin de los centroides.

<p style='text-align: justify;'>Los pasos 2 y 3 se repetirn hasta que los centroides no se muevan o que se muevan debajo de un umbral definido por el investigador. Incluso es posible parar el procedimiento definiendo desde el principio un nmero mximo de repeticiones.</p>

<p style='text-align: justify;'>En el siguiente ejemplo se observa un conjunto de puntos distribuidos, el algoritmo de K-Means buscar generar diferentes conjuntos de puntos, basndose en el posicionamiento de centroides.</p>

## Usando K-Means

```python
from sklearn.cluster import KMeans
```

```python
n_clusters = 2 #numero de clusters objetivo
km = KMeans(n_clusters=n_clusters)
km.fit(normalized_df);
```

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components = 2) #PCA intentar pasar de 17 columnas a 2 dimensiones
X_principal = pca.fit_transform(normalized_df) 
X_principal = pd.DataFrame(X_principal)
X_principal.columns = ['P1', 'P2'] 
X_principal.head()
```

```python
plt.scatter(X_principal['P1'], X_principal['P2'], c = km.predict(normalized_df));
```

## Elbow method o mtodo del codo para saber cuantos clusters elegir

```python
Sum_of_squared_distances = []
for k in range(1,15): #prueba entre 1 y 15 clusters
    km = KMeans(n_clusters=k)
    km = km.fit(normalized_df)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(Sum_of_squared_distances);
#se llama del codo porque el numero de clusters es donde el grafo haga 'un codo'
```

#CONCLUSION: Es ambiguo pero podemos determinar que el optimo es de 4 a 6 clusters; al agregar mas clusters no se obtendra ganancia significativa

## Silhouette method o mtodo de la silueta

```python
from sklearn.metrics import silhouette_score
silhouette_scores = [] 

for n_cluster in range(2, 8):
    silhouette_scores.append( 
        silhouette_score(normalized_df, KMeans(n_clusters = n_cluster).fit_predict(normalized_df))) 
    
# Plotting a bar graph to compare the results 
k = range(2,8)
plt.bar(k, silhouette_scores) 
plt.xlabel('Numbero de clusters', fontsize = 10) 
plt.ylabel('Silhouette Score', fontsize = 10) 
plt.show()

#con este mtodo la barra mas pequea indica el tamao optimo de clusters
```

```python
n_clusters = 4 #determinado por el mtodo del codo y de la silueta
km = KMeans(n_clusters=n_clusters)
km.fit(normalized_df);
plt.scatter(X_principal['P1'], X_principal['P2'], c = km.predict(normalized_df));
```

```python
#Agregamos la etiqueta del cluster correspondiente
df["c"] = km.predict(normalized_df)
df.head()
#la ltima columna 'c' contiene el nmero del cluser al que pertenece cada registro
```
